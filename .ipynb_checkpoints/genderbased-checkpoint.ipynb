```python
# Import our dependencies
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import seaborn as sns
# import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
#  Import and read the heart.csv.
heart_df = pd.read_csv("data.csv")
heart_df

ex_angina={'N': 0,'Y': 1}
heart_df.ExerciseAngina = [ex_angina[item] for item in heart_df.ExerciseAngina]
heart_df
```

```python
#cleaning the dataset by removing all zeroes in the column "cholesterol" as there is no 0 cholesterol.
clean_df=heart_df[heart_df['Cholesterol'] !=0]
clean_df
```

```python
female_df = clean_df[clean_df['Sex'] == 'F']
male_df = clean_df[clean_df['Sex'] == 'M']
```

```python
female_df
```

```python
male_df
```

```python
# There were 390 patients with < 50% diameter narrowing of arteries, and 356 patients with  > 50% diameter narrowing of arteries [ 1 = heart disease, 0 - no heart disease]
male_df['HeartDisease'].value_counts()
female_df['HeartDisease'].value_counts()
```

```python
#Using Oldpeak column from male_df and separate target feature for training data
Oldpeak_df = male_df[['Oldpeak', 'HeartDisease']]
Oldpeak_df
```

```python
def risk (x):
    if x > 1.5 and x < 4.2:
        return 1
    else:
        return 0
    
```

```python
def low (x):
    if x > 0 and x < 2:
        return 1
    else:
        return 0
```

```python
def terrible (x):
    if x ==2.55 and x <2.55:
        return 1
    else:
        return 0
```

```python
def risk (x):
    if x > 1.5 and x < 4.2:
        return 1
    else:
        return 0
    
```

```python
male_df["low"] = male_df["Oldpeak"].apply(low)
male_df.head()
```

```python
male_df["terrible"] = male_df["Oldpeak"].apply(terrible)
male_df.head()
```

```python
male_df["risk"] = male_df["Oldpeak"].apply(risk)
male_df.head()
```

```python
X_train = male_df[["low","risk","terrible"]]
X_test = male_df[["low","risk","terrible"]]
y_train = male_df["HeartDisease"]
y_test = male_df["HeartDisease"]
```

```python
#Convert categorical data to numeric and separate target feature for training and testing data
X_dummies_train = pd.get_dummies(X_train)
X_dummies_train
X_dummies_test = pd.get_dummies(X_test)
X_dummies_test
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

lr = LogisticRegression()
lr.fit(X_dummies_train,y_train)
lr.score(X_dummies_test,y_test)
```

```python
# Scale the data
scaler =  StandardScaler().fit(X_dummies_train)
X_train_scaled = scaler.transform(X_dummies_train)
X_test_scaled = scaler.transform(X_dummies_test)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
lr.fit(X_train_scaled,y_train)
print(lr.score(X_test_scaled ,y_test))
print(lr.score(X_train_scaled,y_train))
```
